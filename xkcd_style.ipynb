```python
data = np.ones(100) # array of ones, 100 of them
print(data.size)
```

```python
data[70:] -= np.arange(30)
#
# what is happening here
# for last 30 elements we are subtracting numbers from the array that strarts from 0 and goes to 29
#
data = np.ones(100)
print(data[70:])
print("np.arange(30)")
print(np.arange(30))
data[70:] -= np.arange(30)
print("so after data[70:] -= np.arange(30) we have data =")
print(data)
```

```python
plt.plot(data)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.xlabel('age')
plt.ylabel('my intellectual capacity')
plt.xticks([]) # no ticks
plt.yticks([]) # no ticks
ax.set_ylim([-30, 30])
plt.plot(data)
plt.title('Whoo Hoo!!!')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.xlabel('age', fontsize=20)
plt.ylabel('my intellectual capacity', fontsize=20)
plt.xticks([])
plt.yticks([])
ax.set_ylim([-30, 30])
plt.plot(data)
plt.title('Whoo Hoo!!!', fontsize = 30)
```
